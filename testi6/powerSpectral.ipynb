```python
#T.18 koodi modattuna, tarkista että kaikki on oikein 


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Data lataaminen ja esikatselu
df = pd.read_csv('Acceleration_data.csv')  # Oletetaan, että tämä on uusi tiedosto
df.head()

plt.figure(figsize=(18,7))
plt.plot(df['seconds_elapsed'], df['z'])  # Päivitetty sarakeviittaukset
plt.xlabel('Aika (s)')
plt.ylabel('Lineaarinen kiihtyvyys z (m/s^2)')
plt.title('Kävelyn jaksollisuuden visualisointi')

# Tehospektrin laskenta
az = df['z'].values  # Valitaan z-komponentti
t = df['seconds_elapsed'].values  # Aika
n = len(az)  # Datapisteiden määrä
T = t[-1] - t[0]  # Aikavälin pituus

# Fourier-muunnos
yf = fft(az)
xf = fftfreq(n, T / n)

# Tehospektri
power_spectrum = np.abs(yf)

# Eniten tehoa sisältävä taajuus
max_power_freq = xf[np.argmax(power_spectrum)]

print(f'Eniten tehoa sisältävä taajuus: {max_power_freq} Hz')

# Suodatusesimerkki: Poistetaan kaikki suuremmat taajuudet kuin eniten tehoa sisältävä taajuus
filtered_signal = np.where(np.abs(xf) > max_power_freq, 0, yf)
filtered_az = np.fft.ifft(filtered_signal)

plt.figure(figsize=(18,7))
plt.plot(t, filtered_az.real)  # Vain reaaliosa kuvataan, koska ifft palauttaa kompleksilukuja
plt.xlabel('Aika (s)')
plt.ylabel('Suodatettu lineaarinen kiihtyvyys z (m/s^2)')
plt.title('Suodatettu signaali')


```
