```python
#Tuodaan kirjastot
import numpy as np
import matplotlib.pyplot as plt

#Kirjastot differentiaaliyhtälöiden ratkaisemiseen
#Analyyttiset ratkaisut (Ratkaistaan yhtälö ja asaadaan yhtälön ratkaisu)

from sympy import Function, dsolve, Eq, Derivative
from sympy.abc import x

#'yhtälö' on annettu numeroina ja ratkaisu on numeroina
#tai sitten alussa ratkaistaan numeeriset arvot analyyttisestä yhtälöstä
from scipy.integrate import odeint

```

```python
#Tehtävä 11 a
#Analyyttinen ratkaisu
#y' + 7y = 0
#y' --> Derivative(y(x), x)
# 7y --> 7*y(x)
y = Function('y') #Määrittelee, että y on funktio
result = dsolve (Derivative(y(x), x) + 7*y(x), y(x)) #Ratkaisee differentiaaliyhtälön
result



```

```python
#Tehtävä 11 a
#Numeerinen ratkaisu
#y' + 7y = 0 --> dydx + 7*y = 0 --> dydx = -7*y

#Määritellään differentiaaliyhtälö Python-funktiona

def f11a(y,x):
    dydx = -7*y
    return dydx

#Määritellään, millä muuttuja x arvoilla haetaan ratkaisua
#Numeerisessa ratkaisussa määritellään rajattu väli x:n arvoja ja ratkaisu on vastaavat y:n arvot

X = np.linspace(0, 10, 101) #Määritellään x:n arvot välillä [0-10], yhteensä 101 pistettä, eli 0.1 välein
#Alkuehto 
y0 = 2 #Määrittelee y:n arvot X:n ensimmäisessä pisteessä y(0) = 2

y = odeint(f11a, y0, X) #Yhtälö, alkuehto, X:n arvot --- Ratkaisee differentiaaliyhtälön numeerisestiy

plt.plot(X, y, '*') #Piirtää X:n ja y:n arvot
plt.xlabel('x') #x-akselin nimi
plt.ylabel('y') #y-akselin nimi
#Piirretään vertailun vuoksi myös analyyttinen ratkaisu
#y = C*e^(-7*x), alkuehdosta saadaan, että y(0) = 2 --> C = 2
plt.plot(X,2*np.exp(-7*X))
plt.plot(X,10*np.exp(-7*X))
plt.legend (['Numeerinen ratkaisu', 'Analyyttinen ratkaisu, kun C = 2, Analyyttinen ratkaisu, kun C = 10'])
```

```python
#Tehtävä 11 b
#y' + cos(x)y = 0
#Analyyttinen ratkaisu
#y' --> Derivative(y(x), x)
#cos(x)*y --> cos(x)*y(x)
from sympy import cos
y = Function('y')

result = dsolve (Derivative(y(x), x) + cos(x)*y(x), y(x)) #dsolve(yhtälö, ratkaistava)
result
```

```python
#Numeerinen ratkaisu
#Määritellään differentiaaliyhtälö Python-funktiona
#Funktion tulee olla muotoa y' = -cos(x)y

#y' --> dydx
#-cos(x)*y --> -np.cos(x)*y
def f11b(y,x):
    dydx = -np.cos(x)*y
    return dydx

X = np.linspace(0, 10, 101) #Määritellään x:n arvot välillä [0-10], yhteensä 101 pistettä, eli 0.1 välein
#Alkuehto 
y0 = 2 #Määrittelee y:n arvot X:n ensimmäisessä pisteessä y(0) = 2

y = odeint(f11b, y0, X) #Yhtälö, alkuehto, X:n arvot --- Ratkaisee differentiaaliyhtälön numeerisestiy

plt.plot(X, y) #Piirtää X:n ja y:n arvot
#Analyyttinen ratkaisu C*np.exp(-np.sin(x)), alkuehto y(0) = 2 --> C = 2
C = 2
plt.plot(X, C*np.exp(-np.sin(X)), '*')
plt.xlabel('x') #x-akselin nimi
C = 3
plt.plot(X, C*np.exp(-np.sin(X)), '*')
plt.ylabel('y') #y-akselin nimi
plt.legend (['Numeerinen ratkaisu', 'Analyyttinen ratkaisu, kun C = 2'])
plt.legend (['Numeerinen ratkaisu', 'Analyyttinen ratkaisu, kun C = 2', 'Analyyttinen ratkaisu, kun C = 3'])


```

```python
#Tehtävä 11 c
#xy' + 3x**2y = 0

#Muokataan muotoon xy' = -3x**2 * y
#--> y' = -3x * y

#Analyyttinen ratkaisu
#y' + 3x*y = 0
#y' --> Derivative(y(x), x)
#3x*y --> 3*x*y(x)

y = Function('y')
result = dsolve (x*Derivative(y(x), x) + 3*x**2*y(x), y(x)) #dsolve(yhtälö, ratkaistava)
result
```

```python
#Numeerinen ratkaisu
# y' = -3x * y

#Määritellään funktio
def f11c(y,x):
    
    dydx = -3*x*y
    return dydx

X = np.linspace(0, 10, 101) #Määritellään x:n arvot välillä [0-10], yhteensä 101 pistettä, eli 0.1 välein
#Alkuehto
y0 = 2 #Määrittelee y:n arvot X:n ensimmäisessä pisteessä y(0) = 2

y = odeint(f11c, y0, X) #Yhtälö, alkuehto, X:n arvot --- Ratkaisee differentiaaliyhtälön numeerisestiy

plt.plot(X, y) #Piirtää X:n ja y:n arvot
#Analyyttinen ratkaisu y = C*np.exp(-3/2*X**2), 
#Alkuehto y(0) = 2 --> C = 2
plt.xlabel('x') #x-akselin nimi
plt.ylabel('y') #y-akselin nimi
C=2
plt.plot(X, C*np.exp(-3/2*X**2))
plt.legend (['Numeerinen ratkaisu', 'Analyyttinen ratkaisu, kun C = 2'])
plt.xlabel('x') #x-akselin nimi
plt.ylabel('y') #y-akselin nimi



```

```python
#Testin 3 tehtävä
#x*y' = 3*y + x**2, y(1) = 0
#Analyyttinen ratkaisu
#Kirjoitetaan muotoon, jossa oikealle vain nolla
#x*y' - 3*y - x**2 = 0
#Kirjoitetaan termit muodossa:
#x*y' --> x*Derivative(y(x), x)
#-3*y --> 3*y(x)
#-x**2 --> -x**2

y = Function('y')
result = dsolve (x*Derivative(y(x), x) - 3*y(x) - x**2, y(x)) #dsolve(yhtälö, ratkaistava)
result
```

Reunaehto y(1) = 0

y(1) = $x^2$ ($C_1$ -1) = 0

Tästä nähdään, että $C_1$ = 1

y(x) = $x^2(x-1)$

y(x) = $x^3 - x^2$

```python
# Differentiaaliyhtälön funktio
def model(y, x):
    dydx = 5*np.exp(-x) - 2*y
    return dydx

# X-akselin arvot
X = np.linspace(0, 10, 101)

# Alkuehto, joka vastaa C = -3 yleisessä ratkaisussa y = 5e^(-x) + Ce^(2x)
y0 = 5 + (-3)

# Ratkaistaan differentiaaliyhtälö numeerisesti
y = odeint(model, y0, X).flatten()

# Piirretään ratkaisu
plt.plot(X, y, label='Numeerinen ratkaisu, kun C = -3')

# Merkitään kuvaajaan tarvittavat tiedot
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
