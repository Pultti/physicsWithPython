```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.03   #sämpläyksen jaksonaika
t = np.arange(0, 20, dt) #määritellään aikavektori
n = len(t)
f1  = 1 #taajuus 1
f2  = 2 #taajuus 2
omega1 = f1 * 2 * np.pi
omega2 = f2 * 2 * np.pi
signal = np.sin(t*omega1) + np.sin(t * omega2) #generoidaan jaksollinen signaali
noisy_signal = signal + 10 * np.random.rand(n) #Lisätään siihen

fig = plt.figure(num=None, figsize= (10, 8), dpi= 100, facecolor= 'w', edgecolor= 'k')
plt.subplot(2, 1, 1)
plt.title('$ \sin(2\pi t) + \sin(4\pi t)$')
plt.plot(t, signal)
plt.subplot(2, 1, 2)
plt.title('Noise added')
plt.plot(t, noisy_signal)
plt.xlabel('t')
plt.show()



```

```python
#lasketaan tehospectri

fft = np.fft.fft(noisy_signal, n) #fourier muunnos
PSD = fft * np.conj(fft) / n #tehospektri
freq = 1/(dt*n) * np.arange(n) #taajuusvektori
L = np.arange(1,np.floor(n/2),dtype = 'int') #rajataan taajuudet positiivisiin

#piirretään kuva
fig = plt.figure(num=None, figsize= (20, 5), dpi= 100, facecolor= 'w', edgecolor= 'k')
plt.plot(freq[L], PSD[L])
plt.title('Power spectral density')
plt.ylabel('Power')
plt.xlabel('Frequency')



plt.show()
```
