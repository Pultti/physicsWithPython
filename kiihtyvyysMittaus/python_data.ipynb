```python
import pandas as pd
df = pd.read_csv('acce.csv.csv',delimiter=';', decimal=',')
import matplotlib.pyplot as plt
```

```python
df = df.rename(columns={"Time (s)": "time", "Linear Acceleration x (m/s^2)": "ax", "Linear Acceleration y (m/s^2)": "ay", "Linear Acceleration z (m/s^2)": "az", "Absolute acceleration (m/s^2)": "a_tot"})
df.info()
df.tail()
```

```python
plt.plot(df.time,df.ax)
plt.plot(df.time,df.ay)
plt.plot(df.time,df.az)
plt.plot(df.time,df.a_tot)
```

```python
# Kuvan määrittely
fig = plt.figure(figsize=(20, 10), dpi=100)

# X-komponentin kuvaaja
plt.subplot(3, 1, 1)
plt.plot(df.time,df.ax)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Linear Acceleration x')
plt.axis([0, 30, -20, 20]) # Olettaen että mittaus kesti 30s ja kiihtyvyysarvot vaihtelevat välillä -20 - 20
plt.legend()

# Y-komponentin kuvaaja
plt.subplot(3, 1, 2)
plt.plot(df.time,df.ay)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Linear Acceleration y')
plt.axis([0, 30, -20, 20])
plt.legend()

# Z-komponentin kuvaaja
plt.subplot(3, 1, 3)
plt.plot(df.time, df.az)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Linear Acceleration z')
plt.axis([0, 30, -20, 20])
plt.legend()

# Näytä kuvaaja
plt.show()
```
